### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
june_info = date_df[date_df.index.month == 6]
december_info = date_df[date_df.index.month == 12]
```

```python
# Identify the average temperature for June
june_avg = june_info['tobs'].mean()

print(f'The average temp in June is {june_avg}')
```

```python
# Identify the average temperature for December
dec_avg = december_info['tobs'].mean()

print(f'The average temp in December is {dec_avg}')
```

```python
# Create collections of temperature data
dec_collection = december_info['tobs']
june_collection = june_info['tobs']
```

```python
# Run paired t-test

pd.set_option('display.float_format', lambda x: '%.5f' % x)
stats.ttest_ind(june_collection, dec_collection)
```

### Analysis

# The low p-value suggests that I can reject the null hypothesis, therefore there is statistical significance in the variance of the two groups of data. Unpaired t-test would be best for this anaysis since we're looking at the means of two independent samples (rather than the means of the same group under separate scenarios).
